```python
import  pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings 
warnings.filterwarnings("ignore")
```

```python
Hr_df=pd.read_csv(r"train_hr.csv",header=0,index_col=0)
Hr_df.head()
```

```python
print(Hr_df.dtypes)
print(Hr_df.shape)
Hr_df.describe()
```

```python
Hr_df.info()
```

```python
Hr_df["education"].fillna(Hr_df["education"].mode()[0],inplace=True)
```

```python
Hr_df["previous_year_rating"]=Hr_df["previous_year_rating"].fillna(np.mean(Hr_df["previous_year_rating"]))
```

```python
Hr_df.info()
```

```python
Hr_df.head()
```

```python
Hr_df=Hr_df.reset_index()
```

```python
Hr_df.head()
```

```python
Hr_df.drop("employee_id",axis=1,inplace=True)
```

```python
Hr_df
```

```python
#Hr_df.drop(Hr_df[Hr_df.is_promoted==0].index,axis=0,inplace=True)
```

```python
Hr_df.shape
```

```python
Hr_df.describe()
```

```python
Hr_df.isnull().sum()
```

```python
Hr_df.columns
```

```python
colname=[]
for x in Hr_df.columns:
    if Hr_df[x].dtype == "object":
        colname.append(x)
colname
```

```python
from sklearn.preprocessing import LabelEncoder

le=LabelEncoder()

for x in colname:
    Hr_df[x]=le.fit_transform(Hr_df[x])
```

```python
X=Hr_df.values[:,:-1]
Y=Hr_df.values[:,-1]

print(X.shape)
print(Y.shape)
```

```python
Hr_df.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()

scaler.fit(X)
X=scaler.transform(X)
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=10)
```

```python
print(X_train.shape)
print(Y_train.shape)
print(X_test.shape)
print(Y_test.shape)
```

# Decision Tree

```python
#predicting using decision_tree_classifier
from sklearn.tree import DecisionTreeClassifier

model_dt=DecisionTreeClassifier(random_state=10,splitter="best") #default value = ("best") & random
#min_samples_leaf, min_samples_split, max_depth, max_features, max_leaf_features

#fix the model on the data and predict the values
model_dt.fit(X_train,Y_train)

Y_pred=model_dt.predict(X_test)
print(Y_pred)
#print(list(zip(Y_test,Y_pred)))
```

```python
model_dt.score(X_train,Y_train) 
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
#confusion_matrix

print(confusion_matrix(Y_test,Y_pred))
print(accuracy_score(Y_test,Y_pred))
print(classification_report(Y_test,Y_pred))
```

# Random Forest

```python
#predicting using decision_tree_classifier
from sklearn.ensemble import RandomForestClassifier

model_rf=RandomForestClassifier(random_state=10,n_estimators=100) #default value = ("best") & random
#min_samples_leaf, min_samples_split, max_depth, max_features, max_leaf_features

#fix the model on the data and predict the values
model_rf.fit(X_train,Y_train)

Y_pred=model_dt.predict(X_test)
print(Y_pred)
#print(list(zip(Y_test,Y_pred)))
```

```python
model_rf.score(X_train,Y_train)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
#create a model
classifier=LogisticRegression()
#fitting training data to the model
classifier.fit(X_train,Y_train)

Y_pred=classifier.predict(X_test)
print(Y_pred)
```

```python
classifier.score(X_train,Y_train)
```
